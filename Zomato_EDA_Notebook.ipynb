```python
#Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
```

```python
#Reading the main dataset

df = pd.read_csv('zomato.csv', encoding = 'latin-1')
df.head()
```

```python
# Fetching the columns of the dataframe

df.columns
```

```python
#Observing the datatypes of the columns

df.info()
```

```python
# Getting descriptive statistics for the columns 

df.describe()
```

## Key activities associated with EDA:

1. Finding missing values
2. Exploring numerical variables
3. Exploring categorical variables
4. Finding relationships between feature variables


```python
#Finding the count of null values in each column

df.isnull().sum()
```

```python
#Getting the column name which has null values

[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#Reading the Country code data

df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#Merging and performing left join with Country Code as the common feature for the Zomato and Country Code dataset respectively.

final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')

final_df.head()
```

Now we will be performing exploration of the features starting with the Country feature.

```python
#Find the count of records by Countries

final_df.Country.value_counts()
```

From the counts of records for Country, we can conclude that most the majority of resturants are of India, which suggests that the base customer and business of Zomato is based on India.

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

### Pie Chart - Top 3 countries


```python
top_3_countries_piechart = plt.pie(country_values[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

#### Country Observation
Zomato's maximum records are from India followed by USA and United Kingdom.

```python
#Get the data related to raings

ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
ratings
```

### Rating Observations

1. When Ratings are between 4.5 to 4.9, it denotes Excellent
2. When Ratings are between 4.0 to 4.4, it denotes Very Good
3. When Ratings are between 3.5 to 3.9, it denotes Good
4. When Ratings are between 2.4 to 3.4, it denotes Average
5. When Ratings are betwwen 1.8 and 2.4, it denotes Poor

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y = 'Rating count', hue = 'Rating color', data = ratings, palette = ['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

### Rating Observations:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 and 3.3

### Count Plot

Count plot is usually used for charting the counts of categorical variable.

```python
sns.countplot(x='Rating color', data = ratings, palette = ['blue', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```

This is not resepct to count of records, but the frequency of the Ratings present in the data.

```python
#Find the countries which have zero ratings
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

### Zero Ratings Observation:
Maximum number of 0 ratings are from Indian records.

```python
#Finding out the currency used in each country of the data

final_df[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Countries which have online deliveries option

final_df[final_df['Has Online delivery'] == "Yes"].Country.value_counts()
```

### Online Delivery Observation:
Online deliveries are available in India and UAE.

```python
## Creating a pie chart for cities distribution

city_values = final_df.City.value_counts().values
city_labels = final_df.City.value_counts().index
```

```python
cities_piechart = plt.pie(city_values[:5], labels = city_labels[:5], autopct = '%1.2f%%')
```

### Cities Observation:
Most the resturants in the dataset are from New Delhi, while remaning are from surrounding cities of New Delhi.


```python
cuisines_values = final_df.Cuisines.value_counts().values
cuisines_labels = final_df.Cuisines.value_counts().index

cusines_chart = plt.pie(cuisines_values[:5], labels = cuisines_labels[:5], autopct='%1.2f%%')
```

### Cuisines Observations:
We find that most of the cusines available in the resturants in the dataset are North Indian cuisine followed by North Indian with Chinese resturants. Only Chinese, Fast Food, North Indian with Muglai have similar share of percentage in the overall cusines chart.
